# Data Preparation

- Scale data [0,1]
- Save each slice in a separate file

```python
import matplotlib.pylab as plt
import numpy as np
import nibabel as nib
import os
import glob
from utils import min_max_normalization, remove_blank_slices
```

```python
data_path = "C:\\Users\\rober\\OneDrive\\Documents\\Datasets\\CC359\\Original\\"
masks_path = "C:\\Users\\rober\\OneDrive\\Documents\\Datasets\\CC359\\STAPLE\\"
domains = ["philips_15", "siemens_15","ge_15"]
```

```python
imgs_list = np.loadtxt("C:/Users/rober/OneDrive/Documents/GitHub/ENSF619.2/SLURM/Unet-Segmentation/Data-split/source_domain_train_validation_set.txt",dtype = str)
print(imgs_list.size)

np.random.shuffle(imgs_list)

train_size = int(0.8*len(imgs_list))
val_size = imgs_list.size - train_size
print(train_size)
print(val_size)

train_set = imgs_list[:train_size]
val_set = imgs_list[train_size:]
```

```python
train_path = "C:/Users/rober/OneDrive/Documents/GitHub/ENSF619.2/SLURM/Unet-Segmentation/Data-preprocessed/Train/"
counter = 1
for ii in train_set:
    ii_mask = ii.split(".nii")[0] + "_staple.nii.gz"
    file_name = os.path.join(data_path,ii + ".gz")
    mask_name = os.path.join(masks_path,ii_mask)
    print(file_name)
    print(mask_name)
    data = nib.load(file_name).get_fdata()
    mask = nib.load(mask_name).get_fdata()
    data_norm = min_max_normalization(data)
    data_cropped,mask_cropped = remove_blank_slices(data_norm, mask)
    H = data_cropped.shape[0]
    for jj in range(H):
        data_slice = data_cropped[jj]
        mask_slice = mask_cropped[jj]
        np.save(os.path.join(train_path,"Images","img_" + str(counter) + ".npy"), data_slice)
        np.save(os.path.join(train_path,"Masks","img_" + str(counter) + ".npy"), mask_slice)
        counter+=1
```

```python
train_path = "C:/Users/rober/OneDrive/Documents/GitHub/ENSF619.2/SLURM/Unet-Segmentation/Data-preprocessed/Val/"
counter = 1
for ii in val_set:
    ii_mask = ii.split(".nii")[0] + "_staple.nii.gz"
    file_name = os.path.join(data_path,ii + ".gz")
    mask_name = os.path.join(masks_path,ii_mask)
    print(file_name)
    print(mask_name)
    data = nib.load(file_name).get_fdata()
    mask = nib.load(mask_name).get_fdata()
    data_norm = min_max_normalization(data)
    data_cropped,mask_cropped = remove_blank_slices(data_norm, mask)
    H = data_cropped.shape[0]
    for jj in range(H):
        data_slice = data_cropped[jj]
        mask_slice = mask_cropped[jj]
        np.save(os.path.join(train_path,"Images","img_" + str(counter) + ".npy"), data_slice)
        np.save(os.path.join(train_path,"Masks","img_" + str(counter) + ".npy"), mask_slice)
        counter+=1
```
